```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()
```

For this parser to work, you need to download the webdriver for your browser for chrome: https://sites.google.com/chromium.org/driver/ or for firefox: https://github.com/mozilla/geckodriver/releases .

In the refs variable put the twitter links you want to parse, the result will be in the twitters variable.

```python
twitters = []

refs = ['https://twitter.com/IBM', 'https://twitter.com/GinniRometty', 'https://twitter.com/JWhitehurst', 'https://twitter.com/arvindkrishna',
        'https://twitter.com/GoldmanSachs',
        'https://twitter.com/Unilever', 
        'https://twitter.com/ServiceNow',
        'https://twitter.com/Boeing', 'https://twitter.com/BoeingAirplanes', 'https://twitter.com/BoeingSpace', 'https://twitter.com/BoeingDefense',
        'https://twitter.com/Sony', 'https://twitter.com/SonyTV', 'https://twitter.com/PlayStation', 'https://twitter.com/SonyPictures', 'https://twitter.com/sonyxperia', 'https://twitter.com/sonymusic',
        'https://twitter.com/Meta',
        'https://twitter.com/nvidia', 'https://twitter.com/NVIDIAGeForce',
        'https://twitter.com/Visa', 
        'https://twitter.com/Mastercard',
        'https://twitter.com/jpmorgan',
        'https://twitter.com/Walmart',
        'https://twitter.com/pfizer',
        'https://twitter.com/CocaCola',
        'https://twitter.com/pepsi', 'https://twitter.com/PepsiCo',
        'https://twitter.com/Adobe', 'https://twitter.com/Photoshop',
        'https://twitter.com/Nike', 'https://twitter.com/nikefootball', 'https://twitter.com/nikestore', 'https://twitter.com/nikebasketball',
        'https://twitter.com/adidas', 'https://twitter.com/adidasoriginals', 'https://twitter.com/adidasfootball', 
        'https://twitter.com/Qualcomm',
        'https://twitter.com/netflix',
        'https://twitter.com/McDonalds', 
        'https://twitter.com/conocophillips',
        'https://twitter.com/PayPal']
        
SCROLL_PAUSE_TIME = 1
    
for ref in refs:    
    driver.get(ref)
    
    twitts = []

    class1 = 'css-1dbjc4n r-j5o65s r-qklmqi r-1adg3ll r-1ny4l3l'
    class2 = 'css-901oao r-18jsvk2 r-37j5jr r-a023e6 r-16dba41 r-rjixqe r-bcqeeo r-bnwqim r-qvutc0'
    class3 = 'css-901oao r-18jsvk2 r-1tl8opc r-a023e6 r-16dba41 r-rjixqe r-bcqeeo r-bnwqim r-qvutc0'

    for k in range(100):
        bs = BeautifulSoup(driver.page_source)

        for j in range(len(bs.find_all('div', {'class' : f"{class1}"}))):
            try:
                bs.find_all('div', {'class' : f"{class1}"})[j].find('div', {'dir' : 'auto', 'class' : [f"{class2}", f"{class3}"]}).get('lang')
            except AttributeError:
                continue
            if bs.find_all('div', {'class' : f"{class1}"})[j].find('div', {'dir' : 'auto', 'class' : [f"{class2}", f"{class3}"]}).get('lang') != 'en':
                continue 
            twit = ''
            for i in range(len(bs.find_all('div', {'class' : f"{class1}"})[j].find('div', {'dir' : 'auto', 'class' : [f"{class2}", f"{class3}"]}).find_all('span'))):
                twit += bs.find_all('div', {'class' : f"{class1}"})[j].find('div', {'dir' : 'auto', 'class' : [f"{class2}", f"{class3}"]}).find_all('span')[i].text
            twitts.append((twit, bs.find_all('div', {'class' : f"{class1}"})[j].find('time').get('datetime')[:-5]))

        # Scroll down
        x = 4500 * (k + 1)
        driver.execute_script(f"window.scrollTo(0, {x})")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)
        
    twitters.append(twitts)
```

```python
accs = ['IBM', 'Ginni Rometty', 'JWhitehurst', 'arvind krishna',
        'GoldmanSachs',
        'Unilever', 
        'ServiceNow',
        'Boeing', 'BoeingAirplanes', 'BoeingSpace', 'BoeingDefense',
        'Sony', 'SonyTV', 'PlayStation', 'SonyPictures', 'sonyxperia', 'sonymusic',
        'Meta',
        'nvidia', 'NVIDIAGeForce',
        'Visa', 
        'Mastercard',
        'jpmorgan',
        'Walmart',
        'pfizer',
        'CocaCola',
        'pepsi', 'PepsiCo',
        'Adobe', 'Photoshop',
        'Nike', 'nikefootball', 'nikestore', 'nikebasketball',
        'adidas', 'adidasoriginals', 'adidasfootball', 
        'Qualcomm',
        'netflix',
        'McDonalds', 
        'conocophillips',
        'PayPal']

df = pd.DataFrame(twitters[0])
df['Twitter'] = 'IBM'

for i in range(1, len(twitters)):
    temp = pd.DataFrame(twitters[i])
    temp['Twitter'] = accs[i]
    df = pd.concat([df, temp], axis=0)
```

```python
df.rename(columns = {0 : 'Text', 1 : 'Time'}, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df['Text'] = df['Text'].apply(lambda x: x.replace('\xa0', ' '))
```

```python
df.drop_duplicates(ignore_index=True, inplace=True)
```

```python
df['Company'] = 1
df.loc[df.query('Twitter in ["IBM", "Ginni Rometty", "JWhitehurst", "arvind krishna"]').index, 'Company'] = 'IBM'
df.loc[df.query('Twitter in ["GoldmanSachs"]').index, 'Company'] = 'GoldmanSachs'
df.loc[df.query('Twitter in ["Unilever"]').index, 'Company'] = 'Unilever'
df.loc[df.query('Twitter in ["ServiceNow"]').index, 'Company'] = 'ServiceNow'
df.loc[df.query('Twitter in ["Boeing", "BoeingAirplanes", "BoeingSpace", "BoeingDefense"]').index, 'Company'] = 'Boeing'
df.loc[df.query('Twitter in ["Sony", "SonyTV", "PlayStation", "SonyPictures", "sonyxperia", "sonymusic"]').index, 'Company'] = 'Sony'
df.loc[df.query('Twitter == "Meta"').index, 'Company'] = 'Meta'
df.loc[df.query('Twitter in ["nvidia", "NVIDIAGeForce"]').index, 'Company'] = 'nvidia'
df.loc[df.query('Twitter == "Visa"').index, 'Company'] = 'Visa'
df.loc[df.query('Twitter == "Mastercard"').index, 'Company'] = 'Mastercard'
df.loc[df.query('Twitter == "jpmorgan"').index, 'Company'] = 'jpmorgan'
df.loc[df.query('Twitter == "Walmart"').index, 'Company'] = 'Walmart'
df.loc[df.query('Twitter == "pfizer"').index, 'Company'] = 'pfizer'
df.loc[df.query('Twitter == "CocaCola"').index, 'Company'] = 'CocaCola'
df.loc[df.query('Twitter in ["pepsi", "PepsiCo"]').index, 'Company'] = 'PepsiCo'
df.loc[df.query('Twitter in ["Adobe", "Photoshop"]').index, 'Company'] = 'Adobe'
df.loc[df.query('Twitter in ["Nike", "nikefootball", "nikestore", "nikebasketball"]').index, 'Company'] = 'Nike'
df.loc[df.query('Twitter in ["adidas", "adidasoriginals", "adidasfootball"]').index, 'Company'] = 'adidas'
df.loc[df.query('Twitter == "Qualcomm"').index, 'Company'] = 'Qualcomm'
df.loc[df.query('Twitter == "netflix"').index, 'Company'] = 'netflix'
df.loc[df.query('Twitter == "McDonalds"').index, 'Company'] = 'McDonalds'
df.loc[df.query('Twitter == "conocophillips"').index, 'Company'] = 'conocophillips'
df.loc[df.query('Twitter == "PayPal"').index, 'Company'] = 'PayPal'
```
